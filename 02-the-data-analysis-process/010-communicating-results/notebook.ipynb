```python
import pandas as pd
%matplotlib inline

df_census = pd.read_csv('../census_income_data.csv')
```

```python
df_a = df_census[df_census['income'] == ' >50K']
df_b = df_census[df_census['income'] == ' <=50K']
```

```python
ind = df_a['education'].value_counts().index
df_a['education'].value_counts()[ind].plot(kind='bar', figsize=(10, 3));
```

```python
df_b['education'].value_counts()[ind].plot(kind='bar', figsize=(10, 3));
```

```python
ind = df_a['workclass'].value_counts().index
df_a['workclass'].value_counts()[ind].plot(kind='pie')
```

```python
df_b['workclass'].value_counts()[ind].plot(kind='pie')
```
